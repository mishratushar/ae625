```python
def error2d(kernel,n0):
    e = np.zeros(n0)
    h = np.zeros_like(e)
    for i in range(n0):
        n = (i+1)*10
        x,y,z,res,h[i] = sph2d(kernel,n)
        e[i] = np.sqrt(np.sum((z-res)**2)/n)
    
    plt.plot(np.log(h),np.log(e))
    slope,intercept=np.polyfit(np.log(h),np.log(e),1)
    print("slope = " ,slope)
```

```python
def gaussian2d_prime(xa,xb,ya,yb,h):
    q = (np.sqrt((xa-xb)**2 + (ya-yb)**2))/h
    scale = 1/(pi*h*h)
    w = scale*np.exp(-(q*q))
    wx = w*(-2)*((xa-xb)/h)
    wy = w*(-2)*((ya-yb)/h)
    for i in range(q.shape[0]):
        for j in range(q.shape[1]):
            if(q[i][j]>3):
                wx[i][j] = 0
                wy[i][j] = 0
    return wx,wy
```

```python
def cubic_spline2d_prime(xa,xb,ya,yb,h):
    q = (np.sqrt((xa-xb)**2 + (ya-yb)**2))/h
```

```python
def grad_w(kernel, N=40, xa=-1, xb=1, ya=-1, yb=1, n=200, k=2, f=0):
    X = np.linspace(xa,xb,n)
    Y = np.linspace(ya,yb,n)

    x,y = np.meshgrid(X,Y)
    zx = pi*np.cos(pi*x)*np.sin(pi*y)
    zy = pi*np.cos(pi*y)*np.sin(pi*x)
    
    dx = (xb-xa)/N
    dy = (yb-ya)/N

    h = k*(dx+dy)
    Xj = np.linspace(-1+dx,1-dx,N)
    Yj = np.linspace(-1+dy,1-dy,N)

    xj,yj = np.meshgrid(Xj,Yj)
    pert = np.random.random((N,N))
    xj += f*pert
    yj += f*pert
    zxj = pi*np.cos(pi*xj)*np.sin(pi*yj)
    zyj = pi*np.cos(pi*yj)*np.sin(pi*xj)

    resx = 0
    resy = 0
    for i in range(N):
        for j in range(N):
            wx,wy = kernel(x,xj[i][j],y,yj[i][j],h)
            resx += dx*dy*wx*zxj[i][j]
            resy += dx*dy*wy*zyj[i][j]
    
    return x,y,zx,zy,resx,resy,h
    
```

```python
def plot2d_der(kernel):
    
    x,y,zx,zy,resx,resy,h = grad_w(gaussian2d_prime)
    fig = plt.figure(1)
    
    ax = plt.axes(projection = "3d")
    ax.plot_surface(x,y,zx,color='b')
    plt.title("True Partial derviative wrt x")
#     plt.xlim(-2.5,2.5)
#     plt.ylim(-2.5,2.5)
#     return

    fig = plt.figure(2)
    ax = plt.axes(projection = "3d")
    ax.plot_surface(x,y,resx,color = 'b')
    plt.title("SPH approximation of Partial derviative wrt x")
#     plt.xlim(-2.5,2.5)
#     plt.ylim(-2.5,2.5)
    
    fig = plt.figure(3)
    ax = plt.axes(projection = "3d")
    ax.plot_surface(x,y,zy,color='g')
    plt.title("True Partial derviative wrt y")
#     plt.xlim(-2.5,2.5)
#     plt.ylim(-2.5,2.5)
    
    fig = plt.figure(4)
    ax = plt.axes(projection = "3d")
    ax.plot_surface(x,y,resy,color='g')
    plt.title("SPH approximation of Partial derviative wrt x")
#     plt.xlim(-2.5,2.5)
#     plt.ylim(-2.5,2.5)
```
